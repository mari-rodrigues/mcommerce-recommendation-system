```python
from nltk.corpus import stopwords
stop_words = stopwords.words('portuguese')
```

```python
# Create the TF-IDF matrix
#tfidf_vectorizer = TfidfVectorizer(stop_words='english')
#tfidf_matrix = tfidf_vectorizer.fit_transform(data['Combined_Text'])

tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words)
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Combined_Text'])
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Combined_Text'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names())
```

```python
# Create the similarity DataFrame using linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
print(np.shape(similarity_matrix))
```

```python
similarity_df = pd.DataFrame(similarity_matrix, index=data['App Name'], columns=data['App Name'])
```

```python
similarity_df
```

```python
similarity_df.to_csv('similarity.csv')
```

```python
def recommend_app(app_name):
    app_index = data[data['App Name'] == app_name].index[0]
    similarity_scores = list(enumerate(similarity_matrix[app_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the most similar apps (excluding the app itself)
    similar_app_indices = [score[0] for score in similarity_scores if data.iloc[score[0]]['Rank'] > 10 and data.iloc[score[0]]['App Name'] != app_name]

    # Get the indices of the top 5 most similar apps
    similar_app_indices = similar_app_indices[:5]

    # Calculate the similarity scores of the top 5 similar apps
    affinity_indices = [score[1] for score in similarity_scores if data.iloc[score[0]]['Rank'] > 10 and score[0] in similar_app_indices]

    # Display the recommended apps with their similarity scores
    recommended_apps = data.iloc[similar_app_indices]
    recommended_apps['Affinity Index'] = affinity_indices
    return recommended_apps[['App Name', 'App Package Name', 'Tier', 'Affinity Index']]

```

```python
# Test the recomendation function
app_name = "SHEIN-Fashion Shopping Online"
print(recommend_app(app_name))
```
