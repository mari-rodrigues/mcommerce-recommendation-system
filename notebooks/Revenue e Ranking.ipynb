```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../csv/All_Investiment.csv')
```

```python
df.info()
```

```python
df = df.groupby(['Advertiser App Name', 'Advertiser App Package Name','Aplicativo_Rank']).agg({'Ad Platforms': 'sum',\
                                                                        'Apps Seen In': 'sum', 'Creatives': 'sum', \
                                                                        'Paid Downloads': 'sum',\
                                                                        'investimento_cpi': 'sum',\
                                                                        'est_numero_anuncios':'sum'}).reset_index()
```

```python
df
```

```python
df.info()
```

```python
investiment = df.groupby(["Advertiser App Name","Advertiser App Package Name",\
                           "Aplicativo_Rank"]).sum()[["investimento_cpi",\
                            "Creatives","Apps Seen In","Paid Downloads","est_numero_anuncios"]].sort_values(by='Aplicativo_Rank',ascending=True).reset_index()
investiment
```

```python
investiment.head(10)
```

```python
investiment['Aplicativo_Rank'] = investiment['Aplicativo_Rank'].astype('category')
```

```python
investiment.info()
```

```python
from scipy.stats import spearmanr
```

```python
x = investiment['Aplicativo_Rank']
y = investiment['investimento_cpi']
corr, p_value = spearmanr(x, y)
corr,p_value
```

```python
# As the app ranking (Aplicativo_Rank) increases, the Investimento_cpi tends to decrease.
# The p-value is extremely small, nearly zero, indicating a statistically significant relationship between the two variables.
```

```python
investiment['Aplicativo_Rank'] = investiment['Aplicativo_Rank'].astype('category').cat.codes

sns.lmplot(data=investiment, x='Aplicativo_Rank', y='investimento_cpi', lowess=True,line_kws={'color':'red'},scatter_kws={'color':'green'})
plt.title("Correlation between Rank and Estimated Investment")
plt.xlabel("Rank")
plt.ylabel("Estimated Investment")
corr, p_value = spearmanr(investiment['Aplicativo_Rank'], investiment['investimento_cpi'])
plt.annotate(f'Spearman Correlation: {corr:.2f}', xy=(0.9, 0.9), xycoords='axes fraction')

plt.legend(["Data"])
```
