```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv('01_General_Shopping_Clean.csv')
df.head()
```

```python
df.info()
```

```python
df['Rank'] = df['Rank'].astype('category')
```

```python
df["Avg Session Duration"] = pd.to_datetime(df["Avg Session Duration"], format="%H:%M:%S")
df["Avg time per User"] = pd.to_datetime(df["Avg time per User"], format="%H:%M:%S")
```

```python
df["Avg time per User"] = df["Avg time per User"].dt.hour.astype(str) + ':' + df["Avg time per User"].dt.minute.astype(str) + ':' + df["Avg time per User"].dt.second.astype(str)
```

```python
df["Avg Session Duration"] = df["Avg Session Duration"].dt.hour.astype(str) + ':' + df["Avg Session Duration"].dt.minute.astype(str) + ':' + df["Avg Session Duration"].dt.second.astype(str)
```

```python
df
```

```python
df.describe().round(3)
```

```python
df_top_10 = df.head(10)
```

```python
df_top_10
```

```python
df_top_10.describe().round(3)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.set_palette('Set2')
sns.barplot(x='Downloads',y='App Name', data=df_top_10)
plt.title('Numero total de downloads dos aplicativos do top 10', fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Total de downloads em milhões', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Nome dos aplicativos', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(np.arange(100000, df_top_10['Downloads'].max()+1, 1000000), rotation=90, ha='right')
plt.show()
```
